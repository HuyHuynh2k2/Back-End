```python
import pandas as pd
import uuid
df = pd.read_csv('books.csv')
```

```python
df1 = pd.DataFrame(columns=['author', 'authorId'])
df2 = pd.DataFrame(columns=['authorId', 'isbn'])
```

```python
for index, row in df.iterrows():
    authors = row['authors'].split(',')
    for author in authors:
        tempdf = pd.DataFrame()
        newRow = {'author' : author}
        new_data = pd.DataFrame([newRow])
        df1 = pd.concat([df1, new_data], ignore_index = True)
        df1 = df1.drop_duplicates()
```

```python
df1['authorId'] = [uuid.uuid4() for _ in range(len(df1.index))]
```

```python
display(df1)
```

```python
for index, row in df.iterrows():
    authors = row['authors'].split(',')
    isbn = row['isbn13']
    for author in authors:
        tempdf = pd.DataFrame()
        tempdf = df1[df1['author'].isin([author])]
        authorId = tempdf['authorId'].to_string()
        authorId = authorId.split()[1]
        newRow = {'authorId': authorId, 'isbn': isbn }
        new_row = pd.DataFrame([newRow])
        df2 = pd.concat([df2, new_row], ignore_index = True)
```

```python
display(df2)
```

```python
df1.to_csv('Author.csv', index=False)
df2.to_csv('BookAuthor.csv', index=False)
```
